```python
# Convert order_time to datetime with a specified format
sales_df['order_time'] = pd.to_datetime(sales_df['order_time'], format='%H:%M:%S', errors='coerce')

# Extract only the time portion, if conversion was successful
sales_df['order_time'] = sales_df['order_time'].dt.time
```

```python
# Fill missing pizza_category based on most frequent category for each pizza_name
sales_df['pizza_category'] = sales_df.groupby('pizza_name')['pizza_category'] \
    .transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)

# Now fill remaining missing pizza_category based on pizza_name_id if available
sales_df['pizza_category'] = sales_df.groupby('pizza_name_id')['pizza_category'] \
    .transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x).where(sales_df['pizza_category'].isnull(), sales_df['pizza_category'])
```

```python
sales_df['pizza_category'].isnull().sum()
```

```python
# Fill missing pizza_category based on most frequent category for each pizza_name
sales_df['pizza_name'] = sales_df.groupby('pizza_category')['pizza_name'] \
    .transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)

# Now fill remaining missing pizza_category based on pizza_name_id if available
sales_df['pizza_name'] = sales_df.groupby('pizza_name_id')['pizza_name'] \
    .transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x).where(sales_df['pizza_name'].isnull(), sales_df['pizza_name'])
```

```python
# Fill missing pizza_category based on most frequent category for each pizza_name
sales_df['pizza_name_id'] = sales_df.groupby('pizza_category')['pizza_name'] \
    .transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)

# Now fill remaining missing pizza_category based on pizza_name_id if available
sales_df['pizza_name_id'] = sales_df.groupby('pizza_ingredients')['pizza_name'] \
    .transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x).where(sales_df['pizza_name_id'].isnull(), sales_df['pizza_name_id'])
```

```python
sales_df['pizza_name_id'].isnull().sum()
```

```python
# Fill missing pizza_category based on most frequent category for each pizza_name
sales_df['pizza_ingredients'] = sales_df.groupby('pizza_category')['pizza_name'] \
    .transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)

# Now fill remaining missing pizza_category based on pizza_name_id if available
sales_df['pizza_ingredients'] = sales_df.groupby('pizza_ingredients')['pizza_name'] \
    .transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x).where(sales_df['pizza_ingredients'].isnull(), sales_df['pizza_ingredients'])
```

```python
sales_df['pizza_ingredients'].isnull().sum()
```

```python
# 1. Calculate total_price if quantity and unit_price are available
sales_df['total_price'] = sales_df['total_price'].fillna(sales_df['quantity'] * sales_df['unit_price'])

# 2. Fill remaining missing total_price based on the average total_price for each pizza_name and pizza_category group
sales_df['total_price'] = sales_df.groupby(['pizza_name', 'pizza_category'])['total_price'] \
                                  .transform(lambda x: x.fillna(x.mean()))
```

```python
Q1 = sales_df['total_price'].quantile(0.25)
Q3 = sales_df['total_price'].quantile(0.75)
IQR = Q3 - Q1
outliers = sales_df[(sales_df['total_price'] < Q1 - 1.5 * IQR) | (sales_df['total_price'] > Q3 + 1.5 * IQR)]
print(outliers)
```

```python
# Calculate IQR for 'quantity' and 'total_price'
for column in ['quantity', 'total_price']:
    Q1 = sales_df[column].quantile(0.25)
    Q3 = sales_df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define outlier thresholds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = sales_df[(sales_df[column] < lower_bound) | (sales_df[column] > upper_bound)]
print(f"Outliers in '{column}':\n", outliers)
```

```python
# Ensure 'order_date' is in datetime format
sales_df['order_date'] = pd.to_datetime(sales_df['order_date'], errors='coerce')

# Group by 'order_date' and 'pizza_category' and sum the 'quantity' column
daily_sales_summary = sales_df.groupby(['order_date', 'pizza_category'])['quantity'].sum().reset_index()

# Display a sample of the summarized data
print(daily_sales_summary.head())
```

```python
# Total sales by pizza category across all days
total_sales_by_category = daily_sales_summary.groupby('pizza_category')['quantity'].sum().sort_values(ascending=False)
print("Total pizza sales by category:\n", total_sales_by_category)

# Visualize total sales by category
plt.figure(figsize=(10, 6))
total_sales_by_category.plot(kind='bar', color='lightcoral')
plt.title('Total Pizza Sales by Category')
plt.xlabel('Pizza Category')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Resample data by week and sum quantities for each pizza category
weekly_sales_summary = daily_sales_summary.copy()
weekly_sales_summary['order_date'] = weekly_sales_summary['order_date'].dt.to_period('W').dt.start_time

# Group by the weekly period and pizza category, summing quantities
weekly_sales_summary = weekly_sales_summary.groupby(['order_date', 'pizza_category'])['quantity'].sum().reset_index()

# Plotting the aggregated weekly data
plt.figure(figsize=(14, 8))
sns.lineplot(data=weekly_sales_summary, x='order_date', y='quantity', hue='pizza_category', marker="o")
plt.title('Weekly Sales Trends by Pizza Category')
plt.xlabel('Week')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.legend(title='Pizza Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()

```

```python
# Get top 5 pizza categories by total quantity sold
top_categories = daily_sales_summary.groupby('pizza_category')['quantity'].sum().nlargest(5).index

# Filter data to include only top categories
top_categories_sales = daily_sales_summary[daily_sales_summary['pizza_category'].isin(top_categories)]

# Plot only the top categories
plt.figure(figsize=(14, 8))
sns.lineplot(data=top_categories_sales, x='order_date', y='quantity', hue='pizza_category', marker="o")
plt.title('Daily Sales Trends for Top Pizza Categories')
plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.legend(title='Pizza Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()
```

```python
# Create a FacetGrid plot by pizza category
g = sns.FacetGrid(daily_sales_summary, col='pizza_category', col_wrap=3, height=4, aspect=1.5, sharey=False)
g.map_dataframe(sns.lineplot, x='order_date', y='quantity', marker="o")
g.set_axis_labels('Date', 'Quantity Sold')
g.set_titles('{col_name} Category')
g.set_xticklabels(rotation=45)
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Daily Sales Trends by Pizza Category')
plt.show()
```
